```python
import faiss
from sentence_transformers import SentenceTransformer
import numpy as np
```

```python
# 1. Load a pre-trained embedding model
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
# 2. Sample documents
documents = [
    "Apple is a fruit",
    "Cats are wonderful pets",
    "Python is a programming language",
    "I love eating bananas",
    "Dogs are loyal animals",
    "Java is used in backend systems"
]
```

```python
# 3. Convert documents to embeddings
doc_embeddings = model.encode(documents)
```

```python
# 4. Create FAISS index
dimension = doc_embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(doc_embeddings)
```

```python
# 5. Search
query = "Tell me about pets"
query_embedding = model.encode([query])
D, I = index.search(query_embedding, k=2)  # Top 2 closest docs
```

```python
# 6. Print results
print("\nQuery:", query)
print("Top matches:")
for idx in I[0]:
    print(f"- {documents[idx]}")
```
