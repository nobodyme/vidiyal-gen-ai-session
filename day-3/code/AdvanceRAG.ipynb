```python
# Dummy text
text = """The quick brown fox jumps over the lazy dog.
The sky is blue. The sun is bright.

The stars shine at night. Nature is beautiful.

Machine learning is transforming industries."""
```

```python
# --- 1. Simple Fixed Length Chunking ---
simple_splitter = CharacterTextSplitter(
    separator = " ",
    chunk_size = 50,     # small for demo
    chunk_overlap = 0
)
simple_chunks = simple_splitter.split_text(text)
print("Simple Fixed Chunks:\n", simple_chunks)
```

```python
# --- 2. Line Break Chunking ---
line_splitter = CharacterTextSplitter(
    separator = "\n\n",   # Double line breaks
    chunk_size = 1000,
    chunk_overlap = 0
)
line_chunks = line_splitter.split_text(text)
print("\nLine Break Chunks:\n", line_chunks)
```

```python
# --- 3. Dot-based Chunking (Sentence Chunking) ---
sentence_splitter = CharacterTextSplitter(
    separator = ". ",
    chunk_size = 1000,
    chunk_overlap = 0
)
sentence_chunks = sentence_splitter.split_text(text)
print("\nDot (.) based Sentence Chunks:\n", sentence_chunks)
```

```python
# --- 4. Overlapping Chunks ---
overlapping_splitter = CharacterTextSplitter(
    separator = " ",
    chunk_size = 50,
    chunk_overlap = 20
)
overlapping_chunks = overlapping_splitter.split_text(text)
print("\nOverlapping Chunks:\n", overlapping_chunks)
```

```python
# --- 5. Adding Metadata Example ---
metadata = {"title": "Nature and Technology", "author": "John Doe", "source": "Sample Book"}
chunks_with_metadata = [{"content": chunk, "metadata": metadata} for chunk in sentence_chunks]

print("\nChunks with Metadata Example:\n", chunks_with_metadata)
```

```python
# --- 6. Create Embeddings for Chunks ---
model = SentenceTransformer('all-MiniLM-L6-v2')  # Small embedding model

# Embedding the sentence chunks
embeddings = model.encode([chunk["content"] for chunk in chunks_with_metadata])

print("\nEmbedding shape:", embeddings.shape)  # Each embedding shape
```

```python
# Search Example (Similarity)
def cosine_similarity(vec1, vec2):
    return np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))

query = "What shines at night?"
query_vec = model.encode(query)

similarities = [cosine_similarity(query_vec, emb) for emb in embeddings]
```

```python
# Find most similar chunk
most_similar_chunk = chunks_with_metadata[np.argmax(similarities)]
print("\nMost Relevant Chunk:\n", most_similar_chunk)
print("\nSource (Metadata):\n", most_similar_chunk['metadata'])
```
